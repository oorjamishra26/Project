```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

data = pd.read_csv("new_dataset.csv")            #Loading the Data
data
```

```python
'''Creating a split function that'll split the dataset into target variable and 
training dataset'''
def create_target_and_predictors(
    data: pd.DataFrame = None, 
    target: str = "estimated_stock_pct"
):
    X = data.drop(columns=[target]).    #These two sets of data will train the supervised ML Models
    y = data[target]
    return X, y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler

'''Creating a function that'll take in the two sets of data
and create a Random Forest Supervised Machine Learning Model
validating it through K Fold Cross Validation and evaluating
its performance over various performance metrics'''

def train_algorithm_with_cross_validation(
    X: pd.DataFrame = None, 
    y: pd.Series = None
):

#Creating a list to store accuracy of each fold
    accuracy = []
    
#Creating a loop to enter into K folds of Cross Validation such that K = 10
    for fold in range (0, 10):

        x = data.drop('estimated_stock_pct', axis = 1)
        y = data['estimated_stock_pct']
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)   #Creating Training and test samples
        
    # Instantiate algorithm and scaler
        model = RandomForestRegressor()
        scaler = StandardScaler()

    # Scale X data, we scale the data because it helps the algorithm to converge
    # and helps the algorithm to not be greedy with large values
        scaler.fit(x_train)
        x_train = scaler.transform(x_train)
        x_test = scaler.transform(x_test)

        model.fit(x_train, y_train)                   #Training the model
        y_predict = model.predict(x_test)             #Generating Predictions on Test Sample

        mse = mean_squared_error(y_test, y_predict)   #Computing Accuarcy
        r2 = r2_score(y_test, y_predict)

        print(f"Mean Squared Error = {mse}")        
        print(f"R2 Score = {r2}")

        mae = mean_absolute_error(y_true=y_test, y_pred=y_predict)
        accuracy.append(mae)
        print(f"Fold {fold + 1}: MAE = {mae:.3f}")

# Finish by computing the average MAE across all folds
    print(f"Average MAE: {(sum(accuracy) / len(accuracy)):.2f}")
```
